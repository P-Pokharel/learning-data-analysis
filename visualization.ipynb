# Visualization

There are two methods of creating the figures - using the matplotlib GlobalAPI and using the OOP approach.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

# Global API (pyplot)

This is most used and popular among the two methods. It generates MATLAB-styled interface.

```python
x = np.arange(-10, 11)
x
```

```python
plt.figure(figsize=(12, 6))
plt.title('Visualizing some data')
plt.plot(x, x**2)
plt.plot(x, -1*(x**2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('Lines plot in separate figure')

plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1*(x**2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X Squared')
```

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

# OOP approach

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(x, x**2, color='green', linewidth=3, marker='o', markersize=8, label='X^2')
axes.plot(x, -1*(x**2), 'b--', label='-X^2')
axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("Plot using OOP")

axes.legend(['X^2', '-X^2'])

fig
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 6))
```

```python
ax4.plot(np.random.randn(50), c='yellow')
ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)
```

# Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2
```

```python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

# Histogram

```python
values = np.random.randn(1000)
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=100, alpha=0.8,
          histtype='bar', color='steelblue',
          edgecolor='green')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```
