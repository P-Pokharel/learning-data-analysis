# Reading data from an external csv file and plotting the information

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/mobile phone price prediction.csv')
```

```python
df.head()
```

```python
df['Price'] = df['Price'].str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.plot()
```

```python
plt.xlabel('Spec Score')
plt.ylabel('Price')
df.plot(kind='scatter', x='Spec_score', y='Price', alpha=0.7, figsize=(10, 6), title='Price vs Spec Score')
```
